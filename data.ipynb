```python

# Dependencies
import pandas as pd
from pathlib import Path 
import json

```

```python
# Read the csv file
data = pd.read_csv("Resources/file.csv")
data
```

```python
# Drop column  "Unnamed:0" and "Date"
clean_df = data.drop(['Unnamed: 0', "Date"], axis = 1)
clean_df

# convert data type of "Transaction_Date" column to datetime data type
clean_df["Transaction_Date"] = pd.to_datetime(clean_df["Transaction_Date"])

# Drop NaN rows
data_df = clean_df.dropna(how='any')
```

```python
# Create a copy of data_df
data_copy = data_df.copy()
# Rename location to customer_location
data_copy.rename(columns={"CustomerID": "customer_id",
                          "Gender":"gender",
                          "Location": "customer_location",
                          
                          "Tenure_Months": "customer_tenure",
                          "Transaction_ID": "transaction_id",
                          "Transaction_Date": "transaction_date",
                          "Product_SKU": "product_id",
                          "Product_Description": "product_description",
                          "Product_Category": "product_category",
                          "Quantity": "quantity",
                          "Avg_Price": "avg_price",
                          "Delivery_Charges": "delivery_charges",
                          "Coupon_Status":"coupon_status",
                          "GST":"gst",                     
                          "Offline_Spend": "instore_spend",
                          "Online_Spend": "online_spend",
                          "Month": "month",
                          "Coupon_Code": "coupon_code",
                          "Discount_pct": "discount_pct"
                          }, inplace=True)
# Display dataframe
data_copy.head()

```

```python
# Create customer dataframe 
customer_df2 = data_copy[["customer_id", "gender", "customer_location", "customer_tenure"]]
# Drop duplicate customer_ids
customer_new_df = customer_df2.drop_duplicates(subset=['customer_id'])
# Display dataframe
customer_new_df.head()
```

```python
# Create product dataframe 
product_df2 = data_copy[["product_id", "product_description", "product_category", "avg_price"]]
# Drop duplicate product_id columns
product_new_df = product_df2.drop_duplicates(subset=['product_id'])
# Display the product dataframe
product_new_df.head()
```

```python
# Create transaction dataframe
transaction_df = data_copy[["transaction_id", "customer_id", "transaction_date", "product_id", "quantity", "delivery_charges", "coupon_status", "coupon_code", "discount_pct", "gst", "month"]]
# Rename index to 'id' to create unique id column
transaction_df.index.names=['id']
# Display transaction dataframe
transaction_df.head()
```

```python
# Export the  DataFrames to CSV files named "customer.csv", "product.csv", "transaction.csv" and save in the output folder
customer_new_df.to_csv("output/customer.csv", encoding='utf8', index=False)
product_new_df.to_csv("output/product.csv", encoding='utf8', index=False)
transaction_df.to_csv("output/transaction.csv", encoding='utf8', index=True)
```
